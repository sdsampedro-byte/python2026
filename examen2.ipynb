3. Escribir una función donde reciba dos numeros enteros que sera el rango de una serie. Los datos se debe ingresar por pantalla. y debe mostrar todos los números primos de ese rango

```python
num1 = int(input("Ingresa el numero de inicio:"))
num2 = int(input("Ingresa el numero de fin:"))
def numeros_primos(inicio, fin):

    for numero in range(inicio, fin + 1):
        
        if numero > 1:            
            es_primo = True 
            
            for i in range(2, numero):
                if numero % i == 0:
                    es_primo = False 
                    break 
           
            if es_primo == True:
                print(numero)

print("Los numeros primos dentro del rango son:")
numeros_primos(num1, num2)
```

4. Escribir un programa en python donde una función reciba una frase y cuente cuantas vocales tiene la frase.

```python
def contar_vocales(frase):
    contador = 0
    vocales = "aeiouáéíóúAEIOUÁÉÍÓÚ"
    
    for letra in frase:
        
        if letra in vocales:
            contador = contador + 1
            
    return contador

texto = input("Escribe una frase: ")
print("El numero total de vocales son:",contar_vocales(texto))
```
