# Estructuras de control
## 1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’


```python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3

if x > 0:
     print('Valor positivo')
elif x == 0:
     print('Valor nulo')
else:
     print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

## **2 Switch**


```python
a = -6
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```python
a = 'l' #week day
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")

```

## Expresión ternaria
* Estructura if-else condensado en una línea.
* Se recomienda usar solo en casos sencillos.

```python
# # # Ejemplo: cálculo del valor absoluto de un número.
x = -3
resultado = x if x >= 0 else -x
print(resultado)
```

```python
val = -3
if val >=0:
    resultado = val
else:
    resultado = -val
    
print(resultado)
```

## Valor booleano
* Todos los objetos en Python tiene inherentemente un valor booleano: *True* o *False*.
* Cualquier número distinto de 0 ó cualquier objeto no vacío tienen valor *True*.
* El número 0, objetos vacíos y el objeto especial *None* tienen valor *False*.

```python
a = -10 #True
b = None #False
if a:
    print("a es True")
if not b:
     print("b es False")
```

```python
c = [1,2,3]
# a c le reasigno None
c = None #0 es igual a vacio a nada
if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
```

```python
operacion=input('Ingrese que operación quiere realizar:')
x=float(input('Ingrese un número'))
y=float(input('Ingrese un número'))

match operacion:
    case 'suma':
        print(x+y)
    case 'resta':
        print(x-y)
    case 'division':
        if y==0:
            print('La división para cero no existe')
        else:
            print (x/y)
    case 'multiplicacion':
        print(x*y)
    case _:
        print('No existe la operación') 
```
