# Estructuras de control
## 1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’


```python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3

if x > 0:
     print('Valor positivo')
elif x == 0:
     print('Valor nulo')
else:
     print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

## 3 Iteración (bucles)
* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: *while* y *for*.

##### Bucle ‘while’
* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos. <br>
Formato general: <br>
while test: # Mientras se cumple la condición <br>
statements # Instrucciones a ejecutar

```python
 # # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]

while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0

while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
    
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

#### Bucle ‘for’
* Permite recorrer los items de una sequencia o un objeto iterable.
* Funciona en strings, listas, tuplas, etc. <br>
Formato general: <br>
for item in objeto: # Asigna los items del objeto a la variable item en cada iteración <br>
statements # Instrucciones a ejecutar

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']
for pelicula in peliculas:
    print(pelicula)
```

• También se usa para iterar un número preestablecido de veces (*counted loops*):

```python
for i in range(10):
    print(i)
```

• *range* es útil en combinación con *len* porque permite acceder a los elementos de una *secuencia*
por posición.


```python
nombre = 'Pablo'

for i in range(len(nombre)):
    print(nombre[i])
```

```python
for letra in nombre:
    print(letra)
```

• Iteración de tuplas:

```python
tuplas = [(1, 2,4), (3, 4), (5, 6)]
for a, b, c in tuplas:
    print(a, b,c)
```

• Iteración de diccionarios. Se iteran las claves:

```python
diccionario = {'a': 1, 'b': 2, 'c': 3}

for key in diccionario:
    print(f"{key} => {diccionario[key]}")
```

* Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos *items*
y *values*.


```python
for key, value in diccionario.items():
    print(f"{key} => {value}")
```

```python
for value in diccionario.values():
    print(value)
```

* La variable *item* en la cabecera del *for* puede ser cualquier expresión que sea válida como
parte izquierda de una asignación convencional.

```python
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
```

1. Realizar un programa en Python donde permita al usuario ingresar un numero entero positivo repititivamente el programa debe continuar mientras se ingrese un numero mayor a cero. Al finalizar el proceso debe mostrarne la suma de los nimeros ingresados.

```python
suma = 0
a = int(input("Ingresa el primer numero: "))

while a > 0:
    suma = suma + a
    a = int(input("Ingresa otro número: "))

print(f"La suma total de los números (positivos) ingresados es: {suma}")
```

2. Crear un programa en Pyhon donde el usuario ingrese un numero por teclado verificar si un numero es primo o no utilizando while

```python
n = int(input("Ingrese un número: "))
if n < 2:
    print("no primo")
else:
    i = 2
    while i*i <= n:
        if n % i == 0:
            print("es número no primo")
            break
        i += 1
    else:
        print("es número primo")
```
