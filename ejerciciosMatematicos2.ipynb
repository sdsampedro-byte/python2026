2. Escribir un programa en python que pida al usuario un  numero entero por teclado y muestre en pantalla si es par o impar.

<img src="img/diagramaFlujo2.jpg">

```python
numero = int(input("Ingrese un número: "))

if numero % 2==0: 
    print("El número es par")
else:
    print("El número es impar")
```
