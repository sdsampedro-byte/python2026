2. Ingresar por teclado la fecha actual DD/MM/AAAA, también ingresar la fecha de nacimiento de una persona y calcular ¿Cuantos años tiene esa persona en años, meses y días.

```python
#Ingresar fecha actual
print("Fecha Actual")
dia = int(input("Ingrese el dia de la fecha actual (DD):"))
mes = int(input("Ingrese el mes de la fecha actual (MM):"))
año = int(input("Ingrese el año de la fecha actual (AAAA):"))

#Ingresar fecha de cumpleaños
print("Fecha de Nacimiento")
dia_n = int(input("Ingrese el dia de nacimiento (dd):"))
mes_n = int(input("Ingrese el mes de nacimiento (mm):"))
año_n = int(input("Ingrese el año nacimiento (aaaa):"))

#Cálculo
print("Cálculo")
Dia = dia - dia_n
Mes = mes - mes_n
Año = año - año_n

# Si hay días negativos → restamos 1 mes y sumamos ~30 días
Día = Dia + 30 * (Dia < 0)
Mes = Mes - 1 * (Dia < 0)

# Si meses negativos → restamos 1 año y sumamos 12 meses
Mes = Mes + 12 * (Mes < 0)
Año = Año - 1 * (Mes < 0)

# Resultado
print("\nResultado aproximado:")
print(Año, "Año,", Mes, "Mes,", Dia, "Dia")
```

3. Elaborar un programa en Python que permita ingresar el radio y la altura de un cilindro y calcular el volumen del cilindro, sabiendo que la formula es V=pi*r2*h

```python
# Programa para calcular el volumen de un cilindro

print("Cálculo del volumen de un cilindro")
print("Fórmula: V = π × r² × h\n")

#Ingreso de datos
Radio = float(input("Ingrese el radio del cilindro (en metros): "))
Altura = float(input("Ingrese la altura del cilindro (en metros): "))

Volumen = 3.1416 * Radio * Radio * Altura

print("\nEl volumen del cilindro es:", Volumen, "m3")
```

4. Elaborar un programa que permita convertir una distacia medida en Metros "m" a Centimetros "cm", milimetros "mm", y kilómetros "km". El usuario debe ingresar el valor en metros y el programa debe mostrar las 3 equivalencias correspondientes.

```python
#Programa de conversión de medidas

Distancia = float(input("Ingresar la Distancia (m):"))

#Conversión
cm = Distancia * 100
mm = Distancia * 1000
km = Distancia / 1000

#Respuesta de conversión de medidas
print("La conversión a centímetros son:", cm, "cm")
print("La conversión a milímetros son:", mm, "mm")
print("La conversión a kilómetros son:", km, "km")
```
