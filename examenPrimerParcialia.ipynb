1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
# Pedir Peso y estatura
Peso = float(input("Ingrese su peso (kg):"))
Estatura = float(input("Ingrese su estatura (m):"))

# Formula
IMC = Peso / (Estatura * Estatura)

print("\nEl IMC es:", IMC)
```

Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$distancia=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y1,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz


```python
import math
a=9
b= math.sqrt(a)
print(b)

# Pedir Coordenadas
x1 = int(input("Ingrese coordenada x1:"))
x2 = int(input("Ingrese coordenada x2:"))
y1 = int(input("Ingrese coordenada y1:"))
y2 = int(input("Ingrese coordenada y2:"))

# Formula
distancia =  math.sqrt((x2-x1)*(x2-x1))+((y2-y1)*(y2-y1))

print("\nLa distancia es:", distancia)
```
